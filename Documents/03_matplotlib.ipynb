## En çok kullanılan grafik türleri
* line
* scatter
* bar
* hist

```python
import numpy as np
x = np.linspace(0,10,100)
x[:10]
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
sozluk = {"USD":27.01,"EUR":28.05,"STR":45}
fig,ax = plt.subplots()
ax.bar(sozluk.keys(),sozluk.values(),);
```

```python
sozluk = {"USD":27.01,"EUR":28.05,"STR":45}
fig,ax = plt.subplots()
ax.barh(list(sozluk.keys()),list(sozluk.values()),);
```

```python
import seaborn as sns
sns.barplot(list(sozluk.keys()),list(sozluk.values()))
```

```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.random(1000)
fig,ax = plt.subplots()

ax.hist(x);
```

```python
x = np.random.random(1000)
sns.distplot(x,kde=True)
```

```python
x = np.random.randn(1000)
sns.displot(x,kde=True)
```
